### Import libraries

```python
import numpy as np
import pandas as pd
```

### Read CSV file

```python
df = pd.read_csv('https://raw.githubusercontent.com/fendiirfan/Hoax_Detection_Microcredential_Associate_Data_Scientist/main/datasetHoax.csv')
```

```python
df.head()
```

```python
df['label'].value_counts()
```

### Classifying hoax based on news title 

```python
X = df['judul']
y = df['label']
```

```python
X.shape
```

```python
y.shape
```

### Spliting training and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
y_train.head()
```

# Machine Learning

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_Vec = vectorizer.fit_transform(X_train)

X_test_Vec = vectorizer.transform(X_test)

print(X_train[0])
print(X_train_Vec[0].toarray())
```

### Building machine learning model

```python
from sklearn import svm

clf_svm = svm.SVC()
clf_svm.fit(X_train_Vec, y_train)
clf_svm.predict(X_test_Vec[0])
```

### Model evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix

Y_pred = clf_svm.predict(X_test_Vec)

print(classification_report(y_test,Y_pred))

cm = confusion_matrix(y_test, Y_pred)

print('Confusion matrix for SVM\n', cm)
```

```python
test_set = ['besok tidak jadi kiamat, pemerintah sipakan penanganan pandemi']
new_test = vectorizer.transform(test_set)

clf_svm.predict(new_test)
```
